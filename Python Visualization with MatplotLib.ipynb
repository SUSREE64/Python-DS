### Python Data Visualization with Matpltolib 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.linspace( 1, 10, 10)
y = x**2
```

### Simple functional way of plottting 

```python
plt.plot(x,y)
plt.show()
```

### Object Oriented Plotting

```python
# Create an object figure first
fig = plt.figure()
# add_axes for the figure the sizes are to be given as floats
axes = fig.add_axes([1.0, 1.0, 1.0, 1.0 ])
axes.plot(x,y)
#Now for the axes we can see methods built it for various cosmetics to the plot
axes.set_xlabel("Xlabel")
axes.set_ylabel('YLabel')
axes.set_title('This is title')
plt.show()
```

### Adding a figure inside a figure with axes

```python
fig = plt.figure()

fig1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
fig2 = fig.add_axes([0.15, 0.45, 0.4, 0.4])

fig1.plot(x,y)

fig2.plot(y,x)

plt.show()
```

###  Subplots side by side

```python
fig, axes = plt.subplots(nrows = 1, ncols = 2)
fig.suptitle('Horizontally stacked subplots')
fig.set_figheight(5)
fig.set_figwidth(15)
axes[0].plot(x,y)
axes[1].plot(y,x)
```

### Subplots vertically stacked

```python
fig, ax = plt.subplots(2)
fig.set_figheight(10)
fig.set_figwidth(15)
fig.suptitle("Vertically Stacked plots")
ax[0].plot(x,y)
ax[1].plot(y,x)
```

### Plotting two Traces on the same fig object with legend

```python
fig = plt.figure(figsize = (5,5))
plt.plot(x,x**3, label = 'X-Cubed')
plt.plot(x,x**2, label = 'X-Squared')

plt.legend()
```

### Plots Styling

```python
fig = plt.figure(figsize = (15,5))
#Adding colour, line width, line style , marker points, markersize, markeredgewidth, markeredgecolor
plt.plot(x,y, lw = 3, color = 'orange', ls = '-', marker= "o", markersize = 20, markerfacecolor = 'yellow',markeredgewidth = 3
        , markeredgecolor = 'green')
plt.title('Simple Line plot')
plt.xlabel("X Label")
plt.ylabel("Y Label")
plt.show()
```
